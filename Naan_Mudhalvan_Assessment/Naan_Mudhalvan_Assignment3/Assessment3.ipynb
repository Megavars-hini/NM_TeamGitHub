```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (replace 'your_dataset.csv' with the actual file path or URL)
data = pd.read_csv('/content/House Price India.csv')

# Univariate Analysis (Histogram for a single variable)
plt.hist(data['number of bedrooms'], bins=20)
plt.xlabel('Variable Name')
plt.ylabel('Frequency')
plt.title('Histogram of Variable Name')
plt.show()

# Bi-Variate Analysis (Scatter plot for two variables)
plt.scatter(data['number of bedrooms'], data['Date'])
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.title('Scatter Plot between Variable 1 and Variable 2')
plt.show()

# Multi-Variate Analysis (Correlation matrix)
correlation_matrix = data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Descriptive statistics of the dataset
summary_stats = data.describe()
print(summary_stats)

# Handle Missing Values (you can customize this based on your data)
# Example: Drop rows with missing values
data.dropna(inplace=True)

# Example: Impute missing values with the mean (replace 'column_name' with the actual column name)
data['number of bedrooms'].fillna(data['Date'].mean(), inplace=True)
```
